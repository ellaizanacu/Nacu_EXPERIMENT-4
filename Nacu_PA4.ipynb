# Problem 1: Create the following DataFrames

## 1. Visayas DataFrame
Create a DataFrame with **Math < 70** and **Hometown = Visayas**:
- We will use `pd.DataFrame()` and Boolean indexing for filtering the data.
- The resulting DataFrame will have the columns `Name`, `Gender`, `Track`, and `Math`.
- Condition: `Math < 70` and hometown is `Visayas`
- Columns: `Name`, `Gender`, `Track`, `Math`

## &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; **EXPECTED OUTPUT:**
   | Name | Gender | Track           | Math |
   |------|--------|-----------------|------|
   | S4   | Male   | Instrumentation | 65   |
   | S11  | Female | Communication   | 48   |
   | S22  | Female | Communication   | 64   |

```python
# Importing the pandas library needed
import pandas as pd

# Loading the corresponding 'board2.xlsx' file given into a data frame named 'data' using pandas
data = pd.read_excel('board2.xlsx')

# Calling the variable 'data' to output the values as a whole
data
```

```python
# Problem 1: DataFrame for Math < 70 and Hometown = Visayas

# Filtering the data for Visayas and Math < 70
visayas_df = data[(data['Hometown'] == 'Visayas') & (data['Math'] < 70)][['Name', 'Gender', 'Track', 'Math']]

# Calling the filtered DataFrame to output the values
visayas_df
```

## a. Instrumentation DataFrame
Create a DataFrame with **Track = Instrumentation**, **Electronics > 70**, and **Hometown = Luzon**:
- We will filter for students who follow the Instrumentation track, and their `Electronics` score is greater than 70.
- The resulting DataFrame will include the columns `Name`, `GEAS`, and `Electronics`.
- Condition: `GEAS > 70`, track is `Instrumentation`, hometown is `Luzon`
- Columns: `Name`, `GEAS`, `Electronics`, `Track`, `Hometown`

## &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp;  **EXPECTED OUTPUT:**

   | S8   | 64   | 81          |
   | S30  | 57   | 81          |

## b. Mindanao Female DataFrame
Create a DataFrame for **Gender = Female**, **Hometown = Mindanao**, and **Average >= 55**:
- We calculate the average score based on the `Math`, `Electronics`, `GEAS`, and `Communication` fields.
- The resulting DataFrame will include the columns `Name`, `Track`, `Electronics`, and `Average`.
- Condition: `Electronics >= 55`, hometown is `Mindanao`, gender is `Female`
- Columns: `Name`, `Track`, `Electronics`, `Average` 


## &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; **EXPECTED OUTPUT:**
   
   | Name | Track            | Electronics  | Average |
   |------|------------------|--------------|---------|
   | S2   | Communication    |      75      |  67.25  |
   | S3   | Instrumentation  |      74      |  72.75  | 
   | S15  | Microelectronics |      41      |  59.00  |
   | S17  | Microelectronics |      79      |  70.50  | 
   | 20   | Instrumentation  |      60      |  66.50  | 


```python
# Problem 1b: DataFrame for Mindanao, Female, and Average >= 55

# Calculating the average of technical subjects
data['Average'] = data[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)

# Filtering the data for Mindanao, Female, and Average >= 55
mindy_df = data[(data['Hometown'] == 'Mindanao') & (data['Gender'] == 'Female') & (data['Average'] >= 55)][['Name', 'Track', 'Electronics', 'Average']]

# Calling the filtered DataFrame to output the values
mindy_df
```

# Problem 2: Visualization
Create visualizations to show how features contribute to average grade:
- We will use the `seaborn` and `matplotlib.pyplot` library to plot bar charts, box plots, scatterplots, and more for comparison.

## **Expected Output**
- Bar chart visualization of average grade by Track, Gender, and Hometown.

```python
# Importing the seaborn and matplotlib.pyplot library
import seaborn as sns
import matplotlib.pyplot as plt

# Setting up the visualization for the contribution of features to get the average grade
plt.figure(figsize=(10, 6))

# Ploting the average grades by Track, Gender, and Hometown
sns.barplot(x='Track', y='Average', hue='Gender', data=data)
plt.title('Average Grade by Track and Gender')

# Calling the values
plt.show()
```

```python
# Alternative visualization for hometown comparison
plt.figure(figsize=(10, 6))
sns.barplot(x='Hometown', y='Average', hue='Track', data=data)
plt.title('Average Grade by Hometown and Track')

# Calling the values
plt.show()
```

```python
# Creating a FacetGrid to show the contribution of Track, Gender, and Hometown to the average grade
g = sns.catplot(
    data=data,
    x='Track',
    y='Average',
    hue='Gender',
    col='Hometown',
    kind='bar',
    height=6,
    aspect=1.2,
    palette='muted')

# Adjusting the title and layout
g.fig.suptitle('Comparison of Average Grades by Track, Gender, and Hometown', y=1.05, fontsize=16)
g.set_axis_labels('College Track', 'Average Grade')
g.add_legend(title='Gender')

# Show the plot
plt.show()
```

```python
# Alternative Visualization: Scatter Plot - Electronics vs Average Grade
# Scatter plot to see the relationship between `Electronics` scores and `Average` grade.
# This plot helps us understand how well students performed in Electronics compared to their average.

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Electronics', y='Average', hue='Gender', style='Track', data=data, s=100)
plt.title('Electronics Scores vs. Average Grade')
plt.show()
```

```python
# Alternative Visualization: Box Plot - Average Grade Distribution by Track
# Box plots can provide a summary of the distribution of average grades across different Tracks.
# This helps to visualize the spread and any potential outliers.

plt.figure(figsize=(10, 6))
sns.boxplot(x='Track', y='Average', hue='Gender', data=data)
plt.title('Distribution of Average Grades by Track')
plt.show()
```
